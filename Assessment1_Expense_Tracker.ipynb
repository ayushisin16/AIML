```python
#1. Add an expense:
#Store the expense as a dictionary.
def createDict(date, category, amount, description):
  Dict = {"Date": date, "Category": category, "Amount": amount, "Description": description}
  return Dict

#Create a function to prompt the user for expense details
def addExpenseDetails(expenses_list):
  date = input("Date of expense (YYYY-MM-DD): ")
  category = input("Category of expense: ")
  amount = input("Amount spent: ")
  description = input("Description of expense: ")
  expense_dict = createDict(date, category, amount, description)
  expenses_list.append(expense_dict)
  print("Expense added:", expense_dict)


#2. View expenses:
'''
  • Function to retrieve and display all stored expenses.
      #o Ensure the function loops through the list of expenses and
      displays the date, category, amount, and description for each entry
  • Validate the data before displaying it
      #o If any required details (date, category, amount, or description) are missing,
       skip the entry or notify the user that it’s incomplete
'''
def retrieveExpenses(expenses):
  total_amount = 0 # Initialize a variable to store the total amount
  if not expenses:
      print("No expenses to display.")
      return 0
  for expense in expenses:
    if "Date" in expense and "Category" in expense and "Amount" in expense and "Description" in expense:
      print("Date:", expense["Date"])
      print("Category:", expense["Category"])
      print("Amount:", expense["Amount"]) # Accessing the 'Amount' key
      print("Description:", expense["Description"])
      print()
      # Assuming the amount is a number, you could add it to a total here
      try:
          total_amount += float(expense["Amount"])
      except ValueError:
          print(f"Warning: Could not convert amount '{expense['Amount']}' to a number.")

    else:
      print("Incomplete entry. Skipping...")
      print()
  return total_amount # Return the calculated total amount


#3. Set and track the budget:
'''• Create a function that allows the user to input a monthly budget. Prompt the user to:
        o Enter the total amount they want to budget for the month
    • Create another function that calculates the total expenses recorded so far
        o Compare the total with the user’s monthly budget
        o If the total expenses exceed the budget, display a warning (Example: You have exceeded your budget!)
        o If the expenses are within the budget, display the remaining balance (Example: You have 150 left for the month)'''

def addMonthlyBudget(current_total_spent):
  budget = input("Enter total amount to budget for the month: ")
  try:
      budget = float(budget)
  except ValueError:
      print("Invalid budget amount. Please enter a number.")
      return # Return None if budget is invalid

  if current_total_spent is not None: # Check if current_total_spent was calculated
      if current_total_spent > budget:
          print("You have exceeded your budget!")
      else:
          remaining = budget - current_total_spent
          print(f"You have {remaining:.2f} left for the month")


#4. Save and load expenses:
import pandas as pd

def saveExpensesToCSV(expenses, filename):
    if not expenses:
        print("No expenses to save.")
        return
    df = pd.DataFrame(expenses)
    df.to_csv(filename, index=False)
    print(f"Expenses saved to {filename}")

def loadExpensesFromCSV(filename):
    try:
        df = pd.read_csv(filename)
        # Convert DataFrame back to a list of dictionaries
        print(f"Expenses loaded from {filename}")
        return df.to_dict(orient = 'records')
    except FileNotFoundError:
        print(f"File '{filename}' not found. Starting with an empty list.")
        return []

#5. Create an interactive menu:
'''• Build a function to display a menu with the following options:
      o Add expense o View expenses o Track budget o Save expenses  o Exit
   • Allow the user to enter a number to choose an option
   • Implement the following conditions:
      o If the user selects option 1, call the function to add an expense
      o If the user selects option 2, call the function to view expenses
      o If the user selects option 3, call the function to track the budget
      o If the user selects option 4, call the function to save expenses to the file
      o If the user selects option 5, save the expenses and exit the program '''

def personalExpenseCalculator():
    all_expenses = loadExpensesFromCSV("expenses.csv") # Load expenses at the start

    while True:
        print("\n--- Personal Expense Calculator Menu ---")
        print("1. Add expense")
        print("2. View expenses")
        print("3. Track budget")
        print("4. Save expenses")
        print("5. Exit")

        userInput = input("Enter the number of the option you want to choose: ")

        if userInput == "1":
            addExpenseDetails(all_expenses)
        elif userInput == "2":
            retrieveExpenses(all_expenses)
        elif userInput == "3":
            current_total_spent = retrieveExpenses(all_expenses) # Calculate total before tracking budget
            addMonthlyBudget(current_total_spent)
        elif userInput == "4":
            saveExpensesToCSV(all_expenses, "expenses.csv")
        elif userInput == "5":
            saveExpensesToCSV(all_expenses, "expenses.csv") # Save before exiting
            print("Exiting Personal Expense Calculator. Goodbye!")
            break # Exit the loop
        else:
            print("Invalid input. Please enter a number between 1 and 5.")

# --- Main part of the script ---
# Start the interactive menu
personalExpenseCalculator()
```
